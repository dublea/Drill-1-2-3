```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams.update({'font.size': 22})
```

```python
hubspot = 'hubspot-customers-2018_08_14.csv'
```

```python
df = pd.read_csv(hubspot)
```

```python
#view columns
df.head()
```

```python
#reduce columns
df_v2 = df.loc[:, ('Name', 'Total Revenue', 'Annual Revenue', 'Industry',
             'Number of Employees', 'Company Domain Name', 'Close Date',
             'First Conversion Date', 'Original Source Type', 'Days to Close',
             'Quantity of Contract UPCs', 'Company owner', 'TrackStreet Customer Number',
             'State/Region', 'Postal Code', 'Is Public', 'Company ID')]
df_v2.head( )
```

```python
tot_rev = np.array(df_v2['Total Revenue'].dropna())
```

```python
sns.distplot(tot_rev, kde=False, rug = True)
```

```python
sns.distplot(tot_rev, kde=False)
```

```python
sns.distplot(tot_rev)
```
